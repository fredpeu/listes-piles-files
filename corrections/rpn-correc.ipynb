# La notation polonaise inversée

Quel est le résultat de l´opération <b>6÷2*(1+2)</b>. 9 ou 1? <br><a href='https://youtu.be/tYf3CpbqAVo'>Cette video</a> pose le problème de l´ambiguité de cette écriture dite <b>infixe</b>.

Généralement, lorsqu’on écrit des expressions mathématiques à l’école, on les écrit de manière **infixe**.<br><br> Par exemple, on écrit 10 - (4 + 3) * 2. **+, * et -** sont des opérateurs infixes.<br> C’est pratique à utiliser car il nous est facile de décomposer cette expression dans notre esprit. L’inconvénient, c’est qu’on a besoin de parenthèses pour indiquer la précédence.

La notation polonaise inverse est une autre façon d’écrire les expressions mathématiques qui n´utilise pas de parenthèse. Bien que la plupart des calculatrices modernes utilisent la notation infixe, certaines personnes ne jurent toujours que par leur calculatrice **RPN** (Reversed Polish Notation). <br><br>Voici ce à quoi l’expression infixe précédente ressemble en RPN : 10 4 3 + 2 * -. Comment calcule-t-on le résultat? <br>Prenons une pile. et parcourons l’expression de gauche à droite. Chaque fois qu’un nombre est rencontré, on l’empile. Dès qu'on rencontre un opérateur, on dépile les deux nombres en sommet de pile, utilisons l’opérateur sur ces deux nombres, et **empilons le résultat**. Si l’expression est bien formée, en arrivant à la fin vous ne devriez plus avoir qu’un nombre dans la pile, et ce nombre est le résultat.<br>

Exemples:

>la forme polonaise de l’expression (21 + 13)  * 40 est: **21  13  +  40  * <br>**
>la forme polonaise de l’expression 10 + (31   * 4 ) est: **10  31  4  *  +<br>**
>la forme polonaise de l’expression  ( 3 + ( 5 * ( 7 + 2 ))) + ( 4 + ( 8 * 9 )) est: <br>  **3  5  7  2  +  *  +  4  8  9  *  +  +<br>**
> Quelles sont les deux formes polonaises de l´expression <b>6÷2*(1+2)</b>?

Réponse:
> 6 1 2 + 2 * /    -> donne **1**<br>
> 6 2 / 1 2 + *    -> donne **9**

Revenons sur le fonctionnement de l´**algorithme d’évaluation**:<br>

Pour effectuer le calcul, chaque donnée numérique est empilée, et lorsqu’on lit un opérateur, les opérandes sont dépilés, l’opération est effectuée et le résultat est mis au sommet de la pile.

Par exemple, pour évaluer  2  3  +  4  * :

>Lecture de 2 qui est reconnu comme donnée numérique et de ce fait est empilé.<br>
>Lecture de 3 qui est également empilé.<br>
>Lecture de + qui est un opérateur binaire : 2 et 3 sont dépilés, le calcul 2 + 3 est effectué, le résultat 5 est empilé.<br>
>Lecture de 4 qui est empilé.<br>
>Lecture de * , opérateur binaire : 4 et 5 sont dépilés, le calcul 4*5 est effectué et le résultat 20 est empilé.<br>
>L’expression est entièrement lue donc le résultat final est au sommet de la pile.<br>

### Elaboration de l´algorithme de calcul:

On cherche à écrire le code de la fonction `rpn`qui calcule le résultat d´une expression en RPN, par exemple:

```python
expression='30  15  7  12  +  *  +  11  8  10  *  +  +'
``` 

Important: **Pour distinguer deux (ou plus) nombres consécutifs, ils sont séparés par des espaces.**

Commençons par écrire la code de la fonction `operateur` qui teste si un caractère `c` est un opérateur arithmétique('+','-','*','/')

````python
def operateur(c):
    if c=='+' or c=='-'or c=='*'or c=='/':
        return True
    else:
        return False
````

Passons à la fonction `calculer` qui calcule le resultat de l'operation *a op b* (avec op = '+','-','*' ou '/'):

````python
def calculer(a,b,op):
    if op=='+':return a+b
    if op=='-':return a-b
    if op=='*':return a*b
    if op=='/':
        if b!=0:return a/b
        else: print("erreur de division par zero")
````

Il ne reste qu´à écrire le code de la fonction `rpn`: <br>
<br>
<img src='http://www.fredpeuriere.com/NSI-term/STRUCTURE/TYPE/rpn.png' width=350/>
<p style="text-align:center;">calcul de: '<b>10 4 3 + 2 * -</b>'</p>

````python
def rpn(expression):
    p=Pile() # creer_pile
    nbr='' 
    for e in expression:
        if operateur(e):# si une opérateur a été trouvée 
            a=p.sommet.element  # on stocke l'élément de sommet de P dans la variable a
            p.depiler()  # on dépile P
            b=p.sommet.element # on stocke l'élément de sommet de P dans la variable b
            p.depiler()
            r=calculer(b,a,e) # on calcule l'opération trouvée
            p.empiler(r) # on empile P avec le résultat              
        else: 
            if e==' ': # si un espace a été trouvé
                if nbr!='': p.empiler(int(nbr)) # on empile le nombre de l'expression 
                nbr='' # initier nbr à vide
            else:
                nbr+=e # construction du nombre             
    return p.sommet.element
expression='6 2 / 1 2 + *'
rpn(expression)
````

Pour simplifier la dernière partie de la fonction, on peut aussi prendre en paramètre une expression sous forme de liste:

````python
def rpn(expression):
    p=Pile() # creer_pile
    for e in expression:
        if operateur(e):# si une opérateur a été trouvée 
            a=p.sommet.element  # on stocke l'élément de sommet de P dans la variable a
            p.depiler()  # on dépile P
            b=p.sommet.element # on stocke l'élément de sommet de P dans la variable b
            p.depiler()
            r=calculer(b,a,e) # on calcule l'opération trouvée
            p.empiler(r) # on empile P avec le résultat              
        else: 
            p.empiler(int(e)) # on empile le nombre de l'expression             
    return p.sommet.element
expression=['6','2','/','1','2','+','*']
rpn(expression)
````
